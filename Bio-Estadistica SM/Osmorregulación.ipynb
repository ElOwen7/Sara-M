# Pruebas de normalidad y homocedasticidad

```python
# Realizar la prueba de normalidad Shapiro-Wilk
normalidad, p_valor_normalidad = shapiro(df3.iloc[:, 1:])
print('Prueba de normalidad Shapiro-Wilk:')
print(f'Estadístico de prueba: {normalidad:.4f}')
print(f'p-valor: {p_valor_normalidad:.4f}')

# Realizar la prueba de homocedasticidad de Levene
homocedasticidad, p_valor_homocedasticidad = levene(*df3.iloc[:, 1:].T.values)
print('\nPrueba de homocedasticidad Levene:')
print(f'Estadístico de prueba: {homocedasticidad:.4f}')
print(f'p-valor: {p_valor_homocedasticidad:.4f}')
```

```python
ns=0.05 #valor predeterminado
while(ns<1 or ns>3):
    ns = float (input(f'Seleccione el nivel de significancia: \n1) 0.05 (5%)\n2) 0.01 (1%)\n3) Especificar valor de significancia.\n')) 
if (ns==1):
    ns=0.05
elif (ns==2):
    ns=0.01
else:
    ns=float(input(f'Ingrese nivel de significancia especifico: '))
    
```

```python
print(f'El nivel de significancia es: {ns}')
```

# Ahora a partir de los resultados anteriores interpretaremos si necesitamos una prueba parametria o no parametrica.

```python
#evaluando normalidad
if (p_valor_normalidad > ns):
    dn = True
else:
    dn = False
```

```python
#evaluando homocedasticidad 
if (p_valor_homocedasticidad > ns):
    dh = True
else:
    dh = False
```

```python
#Determinamos que tipo de prueba se debe o se puede usar
if(dn == True and dh == True):
    print("Los datos siguen una distribución normal y homocedasticidad, entonces se debe usar una prueba de estadística paramétrica")
elif (dn == False and dh == True):
    print("Los datos no siguen una distribución normal y son homocedásticos, entonces se debe usar una prueba de estadística no paramétrica.")    
elif (dn == True and dh == False):
    print("Los datos siguen una distribución normal y no son homocedásticos, entonces se puede usar una prueba de estadística paramétrica o no paramétrica.")    
elif (dn == False and dh == False):
    print("Los datos siguen una distribución normal y no son homocedásticos, entonces se debe usar una prueba de estadística no paramétrica.")    
```

# Prueba no parametrica 

```python
# Seleccionar las columnas de interés (T0, T3, T6, T9, T12, t0, t3, t6, t9, t12 )
columns = ['T0', 'T3', 'T6', 'T9', 'T12','t0', 't3', 't6', 't9', 't12']
data_selected = df3[columns]

# Aplicar la prueba de Friedman
statistic, p_value = friedmanchisquare(*data_selected.values.T)
print("Estadística de Friedman:", statistic)
print("Valor p:", p_value)

```
